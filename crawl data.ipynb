## Danh sách cổ phiếu niêm yết

```python
df_company = listing_companies()
df_company.to_csv('/content/drive/MyDrive/20231/Lưu trữ và xử lý dữ liệu lớn/project/data/list_companies.csv', index=False)
```

## Các chỉ số  tài chính cơ bản

```python
df_ratio = company_fundamental_ratio (symbol=df_company['ticker'][0], mode='simplify', missing_pct=0.8)
n = len(df_company['ticker'])
for i in range(1,n):
  df = company_fundamental_ratio (symbol=df_company['ticker'][i], mode='simplify', missing_pct=0.8)
  df_ratio = pd.concat([df_ratio,df])

df_ratio.to_csv('/content/drive/MyDrive/20231/Lưu trữ và xử lý dữ liệu lớn/project/data/company_fundamental_ratio.csv', index=False)
```

##Dữ liệu giao dịch trong ngày

```python
for i in range(len(df_company)):
    v = df_company['ticker'][i]
    print(v)
    try:
        df_intraday = stock_intraday_data(symbol=v, page_size=1000, investor_segment=False)
        df_intraday.to_csv(f'/content/drive/MyDrive/20231/Lưu trữ và xử lý dữ liệu lớn/project/data/intraday/intraday_{v}.csv', index=False)
    except Exception as e:
        print(f"Error fetching data for {v}: {e}")
```

## Giá lịch sử

```python
n = len(df_company['ticker'])
for i in range(n):
    v = df_company['ticker'][i]
    try:
        df = stock_historical_data(symbol=v, start_date="2013-01-01", end_date="2023-09-29", resolution="1D", type="stock", beautify=True, decor=False, source='DNSE')
        df.to_csv(f'/content/drive/MyDrive/20231/Lưu trữ và xử lý dữ liệu lớn/project/data/history price/hisprice_{v}.csv', index=False)
    except Exception as e:
        print(f"Error fetching historical data for {v}: {e}")
```
